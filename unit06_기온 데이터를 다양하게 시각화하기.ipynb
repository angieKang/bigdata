## 기온 데이터를 히스토그램으로 표현하기(1월과 8월 데이터만)

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)   
result = []  

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '01' : 
            jan.append(float(row[-1]))   
        if month == '08' :
            aug.append(float(row[-1]))
        
plt.hist(jan, bins=100, color='b', label = 'Jan')
plt.hist(aug, bins=100, color='r', label = 'Aug')
plt.legend()
plt.show()
```

## 기온 데이터를 상자 그림으로 나타내기(1월과 8월)

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)   
result = []  

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '01' : 
            jan.append(float(row[-1]))   
        if month == '08' :
            aug.append(float(row[-1]))
        
plt.boxplot(jan)
plt.boxplot(aug)
#plt.boxplot([aug, jan])
plt.show()

#위 아래의 동그라미들은 이상치(outlier)값 표현.
```

## 기온 데이터를 월별로 구분

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)    

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '' :
        #월과 같은 번호의 인덱스에 월별 데이터 저장(예: 1월 -> month[0])
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

## 8월 일별 기온 데이터를 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)    

day = []
for i in range(31):
    day.append([])    #day리스트 내 31개 리스트 생성

for row in data:
    if row[-1] != '' :
        day[int(row[0].split('-')[2])-1].append(float(row[-1]))
        
plt.style.use('ggplot')               #그래프 스타일 지정(배경에 격자)
plt.figure(figsize=(10,5), dpi=300)   #그래프 크기 수정
plt.boxplot(day, showfliers=False)    #아웃라이어 값 생략

plt.show()
```
