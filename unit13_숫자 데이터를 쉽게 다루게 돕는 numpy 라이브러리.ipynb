## pyplot tutorial

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.,5.,0.2) #0.0에서 5.0까지 0.2의 간격으로 가라.
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []

for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
    
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3,'g^')
plt.show()
```

## numpy 라이브러리를 활용해 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
dice = np.random.choice(6,10)

plt.hist(dice, bins = 6)     #dice의 type는 numpy.ndarray
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
dice = []
for i in range(10):
    dice.append(random.randint(1,6))
    
plt.hist(dice, bins = 6)
plt.show()
```

```python
#p속성으로 확률을 설정
import matplotlib.pyplot as plt
import numpy as np
dice = np.random.choice(6, 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])

plt.hist(dice, bins = 6)     
plt.show()
```

```python
# 버블 차트 그리기

import matplotlib.pyplot as plt
import random

x = np.random.randint(10, 100, 200)     #10~100사이의 수 200개 추출
y = np.random.randint(10, 100, 200)
size = np.random.rand(200)*100   #0~100까지의 random

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)      #color의 수는 x의 수, jet 그림으로 표시, alpha는 투명도
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10,100))
    y.append(random.randint(10,100))
    size.append(random.randint(10,100))

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)      #color의 수는 x의 수, jet 그림으로 표시, alpha는 투명도
plt.colorbar()
plt.show()
```

## numpy array 생성하기

```python
#ndarry: N차원 배열
import numpy as np
a = np.array([1,2,3,4])     #list와 같은 연속된 데이터는 array로 가능
print(a)

#리스트 내 요소를 구분하는 쉼표가 없음([1,2,3,4](X), [1 2 3 4](O))
```

```python
#리스트의 특징인 인덱싱과 슬라이싱 가능
import numpy as np
a = np.array([1,2,3,4])
print(a[1], a[-1])     #a의 1번 인덱스 값, -1번 인텍스 값 출력
print(a[1:])           #a의 1번 인덱스를 기준으로 슬라이싱 결과 출력 (0번,1번,2번,3번)
```

```python
#문자타입으로 변환
import numpy as np
a = np.array([1,2,'3',4])
print(a)
```

```python
#배열 초기화 방법
import numpy as np

a = np.zeros(10)    #0으로 이루어진 크기가 10인 배열 생성
print(a)

a = np.ones(10)    #1으로 이루어진 크기가 10인 배열 생성
print(a)

a = np.eye(3)      #3행x3열의 단위 배열 생성
print(a)
```

```python
#연속된 숫자의 array 생성
import numpy as np
print(np.arange(3))          #arange()함수에 1개 값 입력: 3미만의 숫자
print(np.arange(3,7))        #arange()함수에 2개 값 입력: 3이상 7미만의 숫자
print(np.arange(3,7,2))      #arange()함수에 3개 값 입력: 3이상 7미만 숫자 중 2칸씩
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-np.pi, np.pi, np.pi/100)    #pi값을 기준으로 특정 범위 생성, 범위 내 숫자들을 pi값으로 나눈 값만큼 간격
plt.plot(a, np.sin(a))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-np.pi, np.pi, np.pi/100)    
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))     #sin함수의 평행이동(pi만큼 이동)
plt.show()
```

```python
#mask: 어떤 조건에 부합하는 데이터만 선별 저장
import numpy as np
a = np.arange(-5,5)
print(a)

#마스크 적용
print(a<0)        #조건문은 t/f로만 표기됨
print(a[a<0])     #조건에 부합하는 데이터만 남음
#마스크를 변수에 넣어 사용
mask1 = abs(a) > 3     #a배열에 저장된 원소의 절대값이 3보다 크다.
print(a[mask1])        #['-5' '-4' -3 -2 -1 0 1 2 3 '4']
#몇 개의 마스크 연결
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1 + mask2])     #둘 중 하나의 조건이라도 참일 경우
print(a[mask1 * mask2])     #두가지 조건이 모두 참일 경우
```

# numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)     #1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100)*100

mask1 = abs(x) > 50     #x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) > 50

x = x[mask1 + mask2]     #mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2]    

plt.scatter(x, y, s=size, c=x, camp='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
