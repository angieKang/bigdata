## 데이터 읽어와서 최고 기온 데이터 출력하기

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)    #데이터의 맨 윗줄에 있는 헤더 부분을 제외
result = []    #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if(row[-1]) != '' :     #row의 마지막(뒤에서 첫번째)에 있는 최고기온 데이터가 존재한다면
        result.append(float(row[-1]))    #result리스트에 최고 기온 값 추가(문자열->실수형)
#print(result)
#print(len(result))
plt.plot(result,'r')
plt.show()
```

## 날짜 데이터 추출하기

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)    #데이터의 맨 윗줄에 있는 헤더 부분을 제외
result = []    #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '':                        #최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '08' :    #8월에 해당하는 값이라면: row의 첫번째[0](날짜값)을 '-'로 split한 것의 두번째가 8이라면
            result.append(float(row[-1]))    #result리스트에 최고 기온 값 추가
plt.plot(result,'hotpink')                   #result 리스트에 저장된 값을 hotpink색으로 그리기
plt.show()                                   #그래프 나타내기
```

## 날짜 데이터 추출하기(변형)

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)    #데이터의 맨 윗줄에 있는 헤더 부분을 제외
result = []    #최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '':                        
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' :    #2월 14일
            result.append(float(row[-1]))    
plt.plot(result,'hotpink')                   
plt.show() 
```

## 내 생일의 기온 변화를 그래프로 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []    #최고 기온 값을 저장할 리스트 high 생성
low = []     #최저 기온 값을 저장할 리스트 low 생성

for row in data :
    if row[-1] != '' and row[-2] != '' :    #최고 기온 값과 최저 기온 값이 존재한다면
        date = row[0].split('-')
        if 1999 <= int(date[0]) :
            if date[1] == '01' and date[2] == '25' :    #내생년월일 990125
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font', family='Malgun Gothic')    #맑은 고딕을 기본 글꼴로 변경
plt.rcParams['axes.unicode_minus'] = False    #마이너스 기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프')
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
plt.show()
```
