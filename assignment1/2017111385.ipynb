수강생의 생년월일을 입력(input()함수사용)하면 출생년월일 이후부터 2021년 4월 30일까지의 기온 데이터(예를들면, 생년월일.csv)를 이용하여 최고 및 최저 기온을 꺽은선 그래프와 히스토그램으로 표현합니다. 그림에 plt.legend()를 사용해주시고, 출력된 두개의 그래프는 plt.savefig()함수를 사용하여 png파일로 저장합니다.

프로그램 파일(학번.ipynb)과 저장된 그림파일(생년월일_꺽은선.png, 생년월일_히스토그램.png)을 제출합니다.

기온데이터는 출생지의 기온데이터를 기준으로 합니다. https://data.kma.go.kr  기후통계분석>통계분석>기온분석 메뉴로 들어가서 기간은 출생년월일부터 2021년 4월 30일까지로 설정하고 지역은 출생지로 설정한 후 검색 버튼을 누릅니다. 그리고 해당 데이터를 csv파일로 저장한 후 불필요한 부분에 대한 정제를 한후 데이터를 사용합니다.

매년 생년월일에 해당하는 날짜의 최고 기온과 최저 기온을 꺽은선으로 표시하시면 됩니다.

```python
import csv
import matplotlib.pyplot as plt

f = open('nagyeong.csv')
data = csv.reader(f)
next(data)
high = []    #최고 기온 값을 저장할 리스트 high 생성
low = []     #최저 기온 값을 저장할 리스트 low 생성

year, month, day = input('생년월일을 입력해주세요.(ex)1999 01 25): ').split()

for row in data :
    if row[-1] != '' and row[-2] != '' :    #최고 기온 값과 최저 기온 값이 존재한다면
        date = row[0].split('-')
        if year <= date[0]:    #입력한 년도 이후 데이터라면
            if date[1] == month and date[2] == day:    #입력값 날짜일때
                high.append(float(row[-1]))
                low.append(float(row[-2]))
        
        #if year <= date[0] and date[1] <= month and date[2] <= day :     #입력한 출생년월일부터~
        #    high.append(float(row[-1]))
        #    low.append(float(row[-2]))

print(high)
print(low)
        
        
#꺽은선 그래프 그리기
plt.rc('font', family='Malgun Gothic')    #맑은 고딕을 기본 글꼴로 변경
plt.rcParams['axes.unicode_minus'] = False
plt.title(year+month+day+' ~ 20210430')
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
plt.savefig('생년월일_꺽은선.png')    #이미지 파일로 저장
plt.show()

#히스토그램 그리기
plt.title(year+month+day+' ~ 20210430')
plt.hist(high,bins=(len(high)), color='hotpink', label='high')
plt.hist(low,bins=(len(low)), color='skyblue',label= 'low')
plt.legend()
plt.savefig('생년월일_히스토그램.png')    #이미지 파일로 저장
plt.show()
```
