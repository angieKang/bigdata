## 출근 시간대 사람들이 가장 많이 타고 내리는 역은 어디일까

```python
import csv
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
result = []

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    result.append(row[10])     #아침 7시 승차 데이터만 출력해보자

print(len(result))
print(result)
```

```python
#결과를 막대그래프로 출력

import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
result = []

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    result.append(sum(row[10:15:2]))     #아침 7시(10번), 아침 8시(12번), 아침 9시(14번) 승차 인원

result.sort()    #오름차순으로 정렬
plt.figure(dpi = 300)
plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```

```python
#3시간동안 60만명 들어오는 역 찾기

import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    if sum(row[10:15:2]) > mx :     #아침 7시(10번), 아침 8시(12번), 아침 9시(14번) 승차 인원
        mx = sum(row[10:15:2])
        mx_station = row[3] + '('+ row[1] + ')'

print(mx_station, mx)
```

```python
#출근 시간대 사람들이 가장 많이 타고 내리는 역 찾기

import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    a = row[11:16:2]  #하차 인원 값 추출하기
    if sum(a) > mx :     
        mx = sum(a)
        mx_station = row[3] + '('+ row[1] + ')'

print(mx_station, mx)
```

## 밤 11시에 사람들이 가장 많이 타는 역은 어디일까?

```python
#승차시각(t)와 승차인원이 적힌 인덱스(i)의 관계 => i = 4+(t-4)*2

import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : '))

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    a = row[4+(t-4)*2]  #입력 받은 시각의 승차 인원 값 추출하기
    if a > mx :         #모든 데이터 탐색     
        mx = a
        mx_station = row[3] + '('+ row[1] + ')'

print(mx_station, mx)
```

## 시간대별로 사람들이 가장 많이 타고 내리는 역은 어디일까?

```python
# 변수j(0~23까지의 반복을 표현)/ 변수i(승차 인원 값의 인덱스: 4부터 시작)
# i = j*2+4

import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0]*24
mx_station = ['']*24

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    for j in range(24) :
        a = row[j*2+4]         
        if a > mx[j] :         
            mx[j] = a
            mx_station[j] = row[3]

print(mx_station, mx)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0]*24
mx_station = ['']*24

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    for j in range(24) :
        a = row[j*2+4]         
        if a > mx[j] :         
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + '시)'

plt.rc('font', family = 'Malgun Gothic')
plt.figure(dpi = 300)
plt.style.use('ggplot')
plt.bar(range(24), mx)
plt.xticks(range(24),mx_station,rotation=90)
plt.show()
```

```python
#시간대별로 하차 인원이 가장 많은 역을 찾는 코드
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0]*24
mx_station = ['']*24

for row in data:
    row[4:] = map(int,row[4:])    #정수형으로 변환
    for j in range(24) :
        a = row[j*2+5]         
        if a > mx[j] :         
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + '시)'

plt.rc('font', family = 'Malgun Gothic')
plt.figure(dpi = 300)
plt.style.use('ggplot')
plt.bar(range(24), mx, color = 'b')
plt.xticks(range(24),mx_station,rotation=90)
plt.show()
```

## 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면

```python
#1. 데이터를 읽어온다
#2. 모든 역에 대해 시간대별 승차 인원과 하차 인원을 누적해서 더한다.
#3. 시간대별 승차 인원과 하차 인원을 그래프로 표현한다.

import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

s_in = [0]*24
s_out = [0]*24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[i*2+4]
        s_out[i] += row[i*2+5]

plt.figure(dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24),range(4,28))  #24개, 4부터 28까지
plt.show()
```
